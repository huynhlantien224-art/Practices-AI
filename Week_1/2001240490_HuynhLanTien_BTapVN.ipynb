```python
#BT1
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

n1, n2 = len(a), len(b)
min_len = min(n1, n2)

res = [min(a[i], b[i]) for i in range(min_len)]

if n1 > n2:
    res.extend(a[min_len:])
else:
    res.extend(b[min_len:])

print(res)
```

```python
#BT2/a
n = 3
a = [[3, 5, 7],
     [2, 4, 6],
     [7, 8, 9]]

tong = 0
for i in range(n):
    for j in range(n):
        if i + j <= n - 1:
            tong += a[i][j]

print("Tổng tam giác trên đường chéo phụ:", tong)
```

```python
#BT2/b
m, n = 2, 3
a = [[2, -4, -6],
     [-3, -5, 7]]

for i in range(m):
    for j in range(n):
        if a[i][j] < 0:
            a[i][j] = abs(a[i][j])

print("Ma trận sau khi lấy trị tuyệt đối:", a)
```

```python
#BT2/c
m, n = 2, 3
x = 19
a = [[3, 2, 1],
     [4, 5, 7]]

for i in range(m):
    for j in range(n):
        if a[i][j] % 2 == 0:
            a[i][j] = x

print(f"Ma trận sau khi thay số chẵn bằng {x}:", a)
```

```python
#BT2/d
m, n = 2, 2
a = [[1, 4],
     [8, 8]]

toan_chan = True
for i in range(m):
    for j in range(n):
        if a[i][j] % 2 != 0:
            toan_chan = False
            break

print("Ma trận toàn chẵn:", "Đúng" if toan_chan else "Không")
```

```python
#BT2/e
n = 3
a = [[1, 2, 3],
     [2, 5, 4],
     [3, 4, 6]]

doi_xung = True
for i in range(n):
    for j in range(i + 1, n):
        if a[i][j] != a[j][i]:
            doi_xung = False
            break

print("Ma trận đối xứng:", "Đúng" if doi_xung else "Không")
```

```python
#BT2/f
n = 3
a = [[2, 0, 0],
     [0, 3, 0],
     [0, 0, 9]]

tang_dan = True
for i in range(n - 1):
    if a[i][i] >= a[i+1][i+1]:
        tang_dan = False
        break

print("Đường chéo chính tăng dần:", "Đúng" if tang_dan else "Không")
```

```python
#Bai2
#g
n = 3
a = [[1, 2, 5],
     [4, 4, 6],
     [5, 8, 9]]

print("Các phần tử thuộc tam giác dưới đường chéo phụ:")
for i in range(n):
    for j in range(n):
        if i + j >= n - 1:
            print(a[i][j], end=" ")
    print()
```

```python
#Bai2
#h
n = 3
a = [[0, 0, 9],
     [0, 8, 0],
     [1, 0, 0]]

giam_dan = True
for i in range(n - 1):
    if a[i][n - 1 - i] <= a[i+1][n - 2 - i]:
        giam_dan = False
        break

print("Đường chéo phụ giảm dần:", "Đúng" if giam_dan else "Không")
```
