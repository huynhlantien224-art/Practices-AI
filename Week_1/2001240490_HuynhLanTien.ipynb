```python
import random

def tao_mang(m, n):
    a = []
    for i in range(m):
        dong = []
        for j in range(n):
            dong.append(random.randint(0, 99))
        a.append(dong)
    return a
def xuat_mang(a):
    for dong in a:
        for x in dong:
            print(f"{x:4}", end="")
        print()

def xuat_dong_k(a, k):
    print("Dòng", k, ":", a[k])
def xuat_cot_k(a, k):
    print("Cột", k, ":")
    for dong in a:
        print(dong[k])
def tim_dong(a):
    max_tong = -1
    vi_tri = -1
    for i in range(len(a)):
        tong = sum(a[i])
        if tong > 45 and tong > max_tong:
            max_tong = tong
            vi_tri = i
    return vi_tri
def tim_cot(a):
    m = len(a)
    n = len(a[0])
    min_tich = None
    vi_tri = -1

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]

        if min_tich is None or tich < min_tich:
            min_tich = tich
            vi_tri = j

    return vi_tri
def xuat_chan_le(a):
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 != 0:
                print(a[i][j], end=" ")
    print()

def tbc_chan_dong_le(a):
    tong = 0
    dem = 0
    for i in range(len(a)):
        if i % 2 != 0:
            for j in range(len(a[0])):
                if a[i][j] % 2 == 0:
                    tong += a[i][j]
                    dem += 1
    return tong / dem if dem != 0 else 0

def tbc_bien(a):
    m = len(a)
    n = len(a[0])
    tong = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1
    return tong / dem
def tbc_khong_bien(a):
    m = len(a)
    n = len(a[0])

    tong = 0
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tong += a[i][j]
            dem += 1

    if dem == 0:
        return 0
    return tong / dem
m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))

a = tao_mang(m, n)

print("\nMa trận:")
xuat_mang(a)

k = int(input("\nNhập k: "))
xuat_dong_k(a, k)
xuat_cot_k(a, k)

print("\nDòng có tổng > 45 lớn nhất:", tim_dong(a))
print("Cột có tích nhỏ nhất:", tim_cot(a))

xuat_chan_le(a)

print("Trung bình cộng số chẵn dòng lẻ:", tbc_chan_dong_le(a))
print("Trung bình cộng biên:", tbc_bien(a))
print("Trung bình cộng các phần tử không thuộc biên:",tbc_khong_bien(a))
```
